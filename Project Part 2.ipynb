# Step 1

At first, We import torch, math and torchvision from python libraries. We initialized our train data as fashionmnist_train and fashionmnist_test as test data by using FashionMNIST datasets.

```python
import torch
import math
from torchvision import datasets,transforms
import matplotlib.pyplot as plt
fashionmnist_train=datasets.FashionMNIST("../datasets",train=True,download=True,transform=transforms.ToTensor())
fashionmnist_test=datasets.FashionMNIST("../datasets",train=False,download=True,transform=transforms.ToTensor())
```

# Step 2

We imported DataLoader and TensorDataset. By using these libraries, we have Instantiated the dataloaders and checked them if they are iterable or not. 

```python
#dataloader
from torch.utils.data import DataLoader, TensorDataset
print(fashionmnist_train)
dLoader_train = DataLoader(fashionmnist_train,batch_size=256, shuffle=True)
dLoader_test = DataLoader(fashionmnist_test,batch_size=256, shuffle=True)

for batch in dLoader_train:
    print(batch)

for batch in dLoader_test:
    print(batch)
```

# Step 3

To explore our dataset, we used matplotlib library (plt) to show it as an image. Since the linear layer expects each input as a vector, we have to get the elements of the image matrix and we have done this by 'numel' function. This simply gives us the product of the dimensions (28x28x1).

```python
#Exploring the data set
image,label=fashionmnist_train[0]
print(label)
print(image.shape)
plt.imshow(image.squeeze(),cmap='gray')
input_features=image.numel()
input_features #28*28*1

output_features=math.floor(input_features/2)
out_classes=len(fashionmnist_train.classes)
print(fashionmnist_train.classes)
```

# Step 4

Now, we should use the Linear layer of the nn module in torch to define the linear model. But, as we mentioned earlier this layer expects each input instance is a vector so multiple instances are organized in a N, d matrix. At this point, We have to use .Flattend and .Sequential from torch.nn library. We are flattening it into a single-dimensional tensor. It should start flattening to channel dimension (start_dim 1) otherwise if we start at dim 0 each image could be concatenated together since we are flattening the whole batch.

```python
import torch.nn as nn
model=nn.Sequential(nn.Flatten(start_dim=1), 
                    nn.Linear(input_features,out_classes)) #because these will represent our predictions

image=image.unsqueeze(0) #we are adding dimension to indicate it has 1 image in it
print(image.shape)
model(image)
```

# Step 5

We defined the our Cross Entropy Loss function as 'loss' with nn module. 

```python
#Cross Entropy Loss
loss = nn.CrossEntropyLoss()
```

# Step 6

We used 'Stochastic Gradient Descent' as our optimization function.

```python
#Optimization Algorithm
import torch.optim as optim

optimizer=optim.SGD(model.parameters(),0.01)
```

# Step 7

At this step, we defined our training procedure and accuracy function. We also have number of epochs, batch size, and learning rate values as it should be at every training process. In the training process, we updated the model parameters by using our optimizer (SGD). To monitor the training loss evaluation we benefit from matplotlib and printed the loss value after every epoch. In the end, we coded the validation process to evaluate our test data.

```python
#Training
num_epochs=30
bs=256 #generally powers of 2
lr=0.01


#data Loading
dLoader_train = DataLoader(fashionmnist_train,bs, shuffle=True)
dLoader_test = DataLoader(fashionmnist_test,bs, shuffle=True)

#model
model=nn.Sequential(nn.Flatten(start_dim=1), 
                    nn.Linear(input_features,out_classes),
                   )

#Loss
loss = nn.CrossEntropyLoss()

#optimizer
optimizer=optim.SGD(model.parameters(),lr)
    
losses=[]
losses_test=[]
accuracy_train=[]    
accuracy_test=[]
temp_Accuracy=0

#accuracy function
avg_accuracy=0
count=0
def accuracy_func(predictions,target,isAvg=0):
    correct=0
    global avg_accuracy
    global count
    global temp_Accuracy
    if(isAvg==1):
        avg_accuracy=avg_accuracy/count
        return avg_accuracy*100
    
    if(count==7050): #235*30=7050 235 is length of dLoader_train and 30 is epoch number   
        temp_Accuracy=0
        avg_accuracy=0
        count=0
    for i in range(0,len(target),1):
        if(target[i]==predictions[i]):
            correct=correct+1
    accuracy=correct/len(target)
    temp_Accuracy=accuracy
    avg_accuracy+=accuracy
    count=count+1
  
    

#training data
for epoch in range(num_epochs):
    ll=0
    #for training
    for i,(x_batch,y_batch) in enumerate(dLoader_train):
        optimizer.zero_grad()
        y_hat=model(x_batch)
        _,predicted=torch.max(y_hat.data,1)   
        accuracy_func(predicted,y_batch)
        lossVal=loss(y_hat,y_batch)  
        lossVal.backward()
        optimizer.step()
        ll +=lossVal
    print(f'loss training {lossVal}')
    losses.append(ll.item()/len(dLoader_train))
    accuracy_train.append(temp_Accuracy*100)
        
print("Average Accuracy of training data",accuracy_func(0,0,1),"%")
plt.figure(figsize=(16,6))
plt.subplot(1,3,1)
plt.title("training loss")
plt.plot(losses[:])
plt.ylabel('loss')
plt.xlabel('# of epochs')

plt.subplot(1,3,2)
plt.title("training accuracy")
plt.plot(accuracy_train[:])
plt.ylabel('accuracy')
plt.xlabel('# of epochs')


#test data / by using trained model
with torch.no_grad():
    for epoch in range(num_epochs):
        ll_test=0
        for i,(xbatch,ybatch) in enumerate(dLoader_test):
                yhat=model(xbatch)
                _,predictions=torch.max(yhat.data,1)
                accuracy_func(predictions,ybatch)
                lossValtest=loss(yhat,ybatch)
                ll_test+=lossValtest
        print(f'loss test {lossValtest}')
        losses_test.append(ll_test.item()/len(dLoader_test))
        accuracy_test.append(temp_Accuracy*100)
        
    plt.subplot(1,3,3)  
    plt.title("test accuracy")
    plt.plot(accuracy_test[:])
    plt.ylabel('accuracy')
    plt.xlabel('# of epochs')
        
print("Average Accuracy of test data",accuracy_func(0,0,1),"%")        



```
